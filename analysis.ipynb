```python
import pandas as pd
```

```python
file_path = r"D:\insurance\Insurance claims data.csv"

df = pd.read_csv(file_path)

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
```

```python
print("\nFirst 5 rows:")
print(df.head())
```

```python
print("\nInfo:")
print(df.info())

print("\nclaim_status distribution:")
print(df['claim_status'].value_counts())
```

```python
print("\nFirst 5 rows:")
print(df.head())

print("\nInfo:")
print(df.info())

print("\nclaim_status distribution:")
print(df['claim_status'].value_counts())
```

```python
print("\nNumeric summary:")
print(df.describe())

print("\nCategorical summaries:")
for col in ['fuel_type', 'segment', 'engine_type']:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())
```

```python
print("=" * 60)
print("INSURANCE CLAIMS DATASET OVERVIEW")
print("=" * 60)

print(f"\nTotal records: {len(df)}")
print(f"Number of features: {df.shape[1] - 1}")  # excluding target
print(f"Target column: claim_status")
print(f"Non-claim (0): {df['claim_status'].value_counts()[0]}")
print(f"Claim (1): {df['claim_status'].value_counts()[1]}")

print("\nNumeric columns summary:")
print(df.describe().T[['mean', 'min', 'max']])
```

```python
# 1. Claim rate
claim_rate = df['claim_status'].mean() * 100
print(f"Overall claim rate: {claim_rate:.2f}%")

# 2. Claims by fuel type
print("\nClaims by fuel_type:")
print(
    df.groupby('fuel_type')['claim_status']
      .agg(['count', 'mean'])
      .rename(columns={'count': 'total_policies', 'mean': 'claim_rate'})
      .assign(claim_rate=lambda x: (x['claim_rate'] * 100).round(2))
)

# 3. Claims by segment
print("\nClaims by segment:")
print(
    df.groupby('segment')['claim_status']
      .agg(['count', 'mean'])
      .rename(columns={'count': 'total_policies', 'mean': 'claim_rate'})
      .assign(claim_rate=lambda x: (x['claim_rate'] * 100).round(2))
)

# 4. Average ages by claim_status
print("\nAverage subscription_length, vehicle_age, customer_age by claim_status:")
print(
    df.groupby('claim_status')[['subscription_length', 'vehicle_age', 'customer_age']]
      .mean()
      .round(2)
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5,4))
sns.countplot(x='claim_status', data=df)
plt.title('Claim Status Distribution')
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='segment', y='claim_status', data=df, estimator=lambda x: sum(x)/len(x))
plt.title('Claim Rate by Segment')
plt.ylabel('Claim Rate')
plt.show()
```

```python
print("\nOverall claim rate (%):")
print(round(df['claim_status'].mean() * 100, 2))

print("\nClaims by fuel_type:")
claims_by_fuel = (
    df.groupby('fuel_type')['claim_status']
      .agg(['count', 'mean'])
      .rename(columns={'count': 'total_policies', 'mean': 'claim_rate'})
)
claims_by_fuel['claim_rate'] = (claims_by_fuel['claim_rate'] * 100).round(2)
print(claims_by_fuel)

print("\nClaims by segment:")
claims_by_segment = (
    df.groupby('segment')['claim_status']
      .agg(['count', 'mean'])
      .rename(columns={'count': 'total_policies', 'mean': 'claim_rate'})
)
claims_by_segment['claim_rate'] = (claims_by_segment['claim_rate'] * 100).round(2)
print(claims_by_segment)

print("\nAverage subscription_length, vehicle_age, customer_age by claim_status:")
print(
    df.groupby('claim_status')[['subscription_length', 'vehicle_age', 'customer_age']]
      .mean()
      .round(2)
)
```

```python
numeric_cols = ['subscription_length', 'vehicle_age', 'customer_age',
                'region_density', 'airbags', 'displacement',
                'cylinder', 'turning_radius', 'length', 'width',
                'gross_weight', 'ncap_rating', 'claim_status']

corr = df[numeric_cols].corr()
print("\nCorrelation with claim_status:")
print(corr['claim_status'].sort_values(ascending=False))
```
